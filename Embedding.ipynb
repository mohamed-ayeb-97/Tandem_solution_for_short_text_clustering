```python
import pandas as pd
import nltk
nltk.download('punkt')

from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.embeddings.spacy_embeddings import SpacyEmbeddings
```

```python
def clean_and_tokenize_text(input_file_path):
    punctuation = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'

    df = pd.read_csv(input_file_path, header=None)
    df.columns = ["Text"]

    cleaned_texts = []
    tokens = []

    for text in df['Text']:
        cleaned_text = ''.join(char for char in text if char not in punctuation)
        cleaned_texts.append(cleaned_text)

        tokenized_text = ' '.join([word for word in cleaned_text.split() if word not in stop_words])
        tokens.append(tokenized_text)

    df['Text'] = cleaned_texts
    df['clean_text'] = tokens

    return df
```

#importation des données

```python
df_biomed = clean_and_tokenize_text('/content/Biomedical.txt')
```

```python
df_stack = clean_and_tokenize_text('/content/StackOverflow.txt')
```

```python
df_snippet = clean_and_tokenize_text('/content/SearchSnippets.txt')
```

# Embedding

All-MiniLM-L6-V2

```python
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
df_biomed['embeddings']  = df_biomed['clean_text'].apply(lambda x: embeddings.embed_query(x))
```

```python
df_biomed
```

#e5 Transformers Embedding

```python
embeddings2 = HuggingFaceEmbeddings(model_name="intfloat/e5-large")
```

```python
df_stack['embeddings_e5']  = df_stack['tokens'].apply(lambda x: embeddings2.embed_query(x))

df_snippet['embeddings_e5']  = df_snippet['tokens'].apply(lambda x: embeddings2.embed_query(x))

df_biomed['embeddings_e5']  = df_biomed['tokens'].apply(lambda x: embeddings2.embed_query(x))
```

Exportation des embeddings

```python
df_biomed['embeddings_e5'].to_excel('embed_e5_biomed.xlsx', index=False)

df_stack['embeddings_e5'].to_excel('embed_e5_stack.xlsx', index=False)

df_snippet['embeddings_e5'].to_excel('embed_e5_snippet.xlsx', index=False)
```
